#1. Análisis de facturación

```python
ingreso_tienda1 = tienda1['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

print("Ingreso Tienda 1:", ingreso_tienda1)
print("Ingreso Tienda 2:", ingreso_tienda2)
print("Ingreso Tienda 3:", ingreso_tienda3)
print("Ingreso Tienda 4:", ingreso_tienda4)
```

#**Análisis del Desempeño de Tiendas para la Toma de Decisiones Comerciales**
##Introducción
####El análisis de datos se ha convertido en una herramienta clave para apoyar la toma de decisiones estratégicas en organizaciones comerciales. A través del estudio de indicadores de desempeño, es posible identificar oportunidades de mejora y optimizar la gestión de los recursos empresariales (Provost & Fawcett, 2013).

####El objetivo de este análisis es evaluar el rendimiento de cuatro tiendas pertenecientes al Sr. Juan mediante el uso de técnicas de análisis de datos y visualización. Para ello se consideraron diversas variables relevantes para el desempeño comercial, incluyendo los ingresos totales generados por cada tienda, las categorías de productos más vendidas, las calificaciones promedio de los clientes, los productos con mayor volumen de ventas y otros factores relacionados con la experiencia de compra.

####El propósito final del análisis es identificar cuál de las tiendas presenta el menor rendimiento relativo y, en consecuencia, determinar cuál debería ser vendida para mejorar la eficiencia del negocio.


#**Metodología**
####Para el desarrollo del análisis se utilizó Python como herramienta de procesamiento de datos, empleando las bibliotecas Pandas para el análisis de datos y Matplotlib para la visualización gráfica.

####El análisis se estructuró en tres dimensiones principales observadas en los gráficos generados:

- Ingresos totales por tienda

- Distribución de categorías de productos

- Calificación promedio de los clientes

####Estas métricas permiten evaluar tanto el desempeño financiero como la percepción del cliente respecto a cada tienda.

#**Resultados**
##Ingresos totales por tienda
####El análisis de ingresos muestra diferencias claras en el desempeño económico de las tiendas. De acuerdo con el gráfico generado, la Tienda 1 presenta los mayores ingresos, seguida por la Tienda 2 y la Tienda 3. En contraste, la Tienda 4 presenta el menor nivel de ingresos, lo que indica una menor capacidad de generación de ventas en comparación con las demás tiendas.

####Este indicador es particularmente relevante, ya que los ingresos reflejan directamente el aporte de cada tienda al rendimiento general del negocio.

##Distribución de categorías de productos
####El análisis de categorías de productos, representado mediante un gráfico circular para la Tienda 1, muestra que las ventas se distribuyen principalmente entre las categorías de Muebles, Electrónicos, Juguetes y Electrodomésticos, con una participación relativamente equilibrada.

####Esto sugiere que la tienda mantiene una oferta diversificada de productos, lo que puede contribuir a mantener un flujo constante de ventas en diferentes segmentos del mercado.
##Calificación promedio de los clientes

####La satisfacción del cliente es otro indicador importante para evaluar el desempeño de una tienda. El análisis de las calificaciones promedio muestra que todas las tiendas presentan evaluaciones similares cercanas a 4 puntos, lo que indica un nivel de satisfacción general relativamente alto.

####Sin embargo, la Tienda 3 presenta la mejor calificación promedio, lo que sugiere una experiencia de cliente ligeramente superior en comparación con las demás tiendas.



#Discusión

####Al analizar conjuntamente los indicadores observados en los gráficos, se puede identificar que el principal factor diferenciador entre las tiendas es el nivel de ingresos generados. Mientras que las tiendas 1, 2 y 3 mantienen un desempeño económico relativamente alto, la Tienda 4 muestra un rendimiento significativamente menor en términos de ingresos.

####Por otra parte, las calificaciones de los clientes son bastante similares entre todas las tiendas, lo que indica que la satisfacción del cliente no representa una debilidad crítica en ninguna de ellas.

####Por lo tanto, el factor más determinante para la toma de decisión es el rendimiento económico comparativo entre las tiendas.

#Conclusión y recomendación

####Con base en los resultados obtenidos a partir del análisis de datos y las visualizaciones generadas, se concluye que la Tienda 4 presenta el menor desempeño general dentro del conjunto de tiendas analizadas.

####Aunque las calificaciones de los clientes son similares entre todas las tiendas, la Tienda 4 genera los menores ingresos totales, lo que indica una menor contribución al rendimiento global del negocio.

####En consecuencia, se recomienda que el Sr. Juan considere vender la Tienda 4, ya que esta representa la unidad con menor desempeño económico dentro de su portafolio comercial. Esta decisión permitiría concentrar recursos y esfuerzos en las tiendas con mayor capacidad de generación de ingresos y mejor posicionamiento dentro del mercado.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

plt.scatter(tienda1['lon'], tienda1['lat'], alpha=0.3, label='Tienda 1')
plt.scatter(tienda2['lon'], tienda2['lat'], alpha=0.3, label='Tienda 2')
plt.scatter(tienda3['lon'], tienda3['lat'], alpha=0.3, label='Tienda 3')
plt.scatter(tienda4['lon'], tienda4['lat'], alpha=0.3, label='Tienda 4')

plt.title("Distribución geográfica de ventas")
plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.legend()

plt.show()
```

```python
import pandas as pd

todas_ventas = pd.concat([tienda1, tienda2, tienda3, tienda4])
```

```python
plt.figure(figsize=(10,6))

plt.hexbin(
    todas_ventas['lon'],
    todas_ventas['lat'],
    gridsize=40,
    cmap='Reds'
)

plt.colorbar(label='Cantidad de ventas')
plt.title("Mapa de calor de concentración de ventas")
plt.xlabel("Longitud")
plt.ylabel("Latitud")

plt.show()
```
